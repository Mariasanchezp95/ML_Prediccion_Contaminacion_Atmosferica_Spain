```python
# 1. CARGA DE TODOS LOS ARCHIVOS PARQUET DE LA CARPETA E1a

import pandas as pd      # Librería principal para manejo de datos
import glob              # Permite buscar archivos usando patrones (ej: *.parquet)

# -------------------------------------------------------------
# 1) Cargar todos los archivos .parquet dentro de la carpeta E1a
# -------------------------------------------------------------
# glob.glob busca todos los archivos que coinciden con el patrón "*.parquet"
# Esto devuelve una lista con las rutas completas de cada archivo
rutas_archivos = glob.glob(
    r"C:/Users/maria/Documents/GitHub/ML Proyect/src/data_totales/ParquetFiles/E1a/*.parquet"
)

# -------------------------------------------------------------
# 2) Leer cada archivo parquet y concatenarlos en un único DataFrame
# -------------------------------------------------------------
# pd.read_parquet lee cada archivo individualmente
# La lista por comprensión crea una lista de DataFrames
# pd.concat une todos los DataFrames en uno solo
df = pd.concat(
    [pd.read_parquet(f) for f in rutas_archivos],
    ignore_index=True     # Reinicia los índices para evitar duplicados
)

# -------------------------------------------------------------
# 3) Mostrar las primeras filas para comprobar que todo está correcto
# -------------------------------------------------------------
df.head()
```
